```python
ratings=ratings[ratings['user_id'].isin(new_users)]
```

```python
ratings_with_bookname=ratings.merge(books,on='ISBN')
```

```python
ratings_with_bookname.shape
```

```python
ratings_with_bookname.head()
```

```python
book_num_ratings=ratings_with_bookname.copy()
```

```python
book_num_rating=ratings_with_bookname.groupby('title')['rating'].count().reset_index()
```

```python
book_num_rating=book_num_rating[book_num_rating['rating']>=50]
```

```python
book_num_ratings=book_num_ratings[book_num_ratings['title'].isin(book_num_rating['title'])]
```

```python
book_num_ratings
```

```python
book_num_ratings=book_num_ratings.groupby('title')['rating'].mean().reset_index()
```

```python
book_num_ratings.rename(columns={'rating':'avg_ratings'},inplace=True)
```

```python
book_num_ratings
```

```python
book_num_ratings=book_num_ratings[book_num_ratings['avg_ratings']>=1]
```

```python
book_num_ratings
```

```python
final_rating=ratings_with_bookname.merge(book_num_ratings,on='title')
```

```python
final_rating.shape
```

```python
final_rating=final_rating.drop_duplicates(subset=['user_id','title'])
```

```python
final_rating.shape
```

```python
final_rating
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot=book_pivot.fillna(0)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_matrix=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(n_neighbors=10, algorithm='auto',leaf_size=50,radius=5.0)
```

```python
model.fit(book_matrix)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[1,:].values.reshape(1,-1),n_neighbors=10)
```

```python
suggestion
```

```python
def recommend(bookname):
    b_id=np.where(book_pivot.index==bookname)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[b_id-1,:].values.reshape(1,-1),n_neighbors=10)
    for i in suggestion[0]:
        print(book_pivot.index[i])
```

```python
recommend('1984')
```

```python
recommend('You Belong To Me')
```
