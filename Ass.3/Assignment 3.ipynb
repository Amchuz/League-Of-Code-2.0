```python
book_tags = pd.read_csv('book_tags.csv', encoding = "ISO-8859-1")
```

```python
tags = pd.read_csv('tags.csv')
```

```python
lang=['eng','en-US']
books=books[books['language_code'].isin(lang)]
```

```python
book_tags.columns
```

```python
tags.columns
```

```python
book_with_tags = pd.merge(book_tags, tags, on='tag_id', how='inner')
```

```python
book_with_tags
```

```python
ratings=ratings.rename(columns={'book_id':'id'})
```

```python
books_with_ratings=pd.merge(books, ratings, on='id', how='inner')
```

```python
books_with_ratings
```

```python
books_with_ratings=books_with_ratings[books_with_ratings.rating > 3]
```

```python
books_with_ratings
```

```python
books_with_ratings.drop(['work_id','books_count','isbn13','ratings_count','work_ratings_count','work_text_reviews_count','ratings_1','ratings_2','ratings_3','ratings_4','ratings_5','image_url','small_image_url','title','av'],axis=1,inplace=True)
```

```python
tf1 = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix1 = tf1.fit_transform(book_with_tags['tag_name'].head(10000))
cosine_sim1 = linear_kernel(tfidf_matrix1, tfidf_matrix1)
```

```python
titles1 = books['title']
indices1 = pd.Series(books.index, index=books['title'])

# Function that get book recommendations based on the cosine similarity score of books tags
def tags_recommendations(title):
    idx = indices1[title]
    sim_scores = list(enumerate(cosine_sim1[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    book_indices = [i[0] for i in sim_scores]
    return titles1.iloc[book_indices]
```
