```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('music.csv',error_bad_lines=False,encoding='latin-1')
```

```python
df
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df
```

```python
df.drop(['bpm','nrgy','dnce','dB','live','val','dur','acous','spch','pop'],axis=1,inplace=True)
```

```python
df
```

```python
df.rename(columns={'top genre':'genre'},inplace=True)
```

```python
df
```

```python
l=df['genre'].value_counts()
```

```python
l
```

```python
df['count'] = df.groupby('genre')['genre'].transform('count')
```

```python
df
```

```python
df=df[df['count']>=10]
```

```python
df
```

```python
l
```

```python
df["genre"].replace({"dance pop": 1,"pop": 2,"canadian pop": 3,"barbadian pop": 4, "boy band": 5, "electropop": 6,"british soul": 7,"big room": 8,}, inplace=True)
```

```python
df
```

```python
df_pivot=df.pivot_table(columns='artist',index='title',values='genre')
```

```python
df_pivot
```

```python
df_pivot=df_pivot.fillna(0)
```

```python
df_pivot
```

```python
from scipy.sparse import csr_matrix
df_matrix=csr_matrix(df_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(n_neighbors=10, algorithm='auto',leaf_size=50,radius=5.0)
```

```python
model.fit(df_matrix)
```

```python
distance,recommendation=model.kneighbors(df_pivot.iloc[1,:].values.reshape(1,-1),n_neighbors=10)
```

```python
recommendation
```

```python
def moresongs(songname):
    sid=np.where(df_pivot.index==songname)[0][0]
    distance,recommendation=model.kneighbors(df_pivot.iloc[sid-5,:].values.reshape(1,-1),n_neighbors=10)
    for i in recommendation[0]:
        print(df_pivot.index[i])
```

```python
moresongs('Younger Now')
```

```python
moresongs('Closer')
```
